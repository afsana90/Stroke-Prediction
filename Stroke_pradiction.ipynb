```python
data=data.drop(['id'],axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.fillna(method='ffill',inplace=True)
data.isnull().sum()
```

```python
data_encoded = pd.get_dummies(data, columns=['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'])#,drop_first=True)
data_encoded = data_encoded.astype(int)
print(data_encoded.head())
#data_encoded.info()
#data_encoded.dtypes
```

```python
print("Encodeed DataFrame:")
print(data_encoded.head())
```

```python
data.hist(bins=50,figsize=(12,10))
plt.show()
```

```python
data.corr(numeric_only=True)
sns.heatmap(data.corr(numeric_only=True),annot=True)
plt.show()
```

```python
target_name='stroke'
y = data[target_name]
x = data_encoded.drop(target_name, axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

Random Forest Algorithm


```python
from sklearn. ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
print("Train Accuracy of Random Forest Algorithm ", rf.score(x_train,y_train)*100)
print("Test Accuracy score of Random Forest Algorithm ", rf.score(x_test,y_test)*100)
```

```python
prediction = rf.predict([[int(x) for x in ["50","2","168","62","110","80","1","1","0","0","1"]] + [0] * 10])
if(prediction !=0):
  print("The person has stroke")
else:
  print("The person has no stroke")
```

```python
prediction = rf.predict([[int(x) for x in ["55","1","156","85","140","90","3","1","0","0","1"]] + [0] * 10])
if(prediction !=0):
  print("The person has stroke")
else:
  print("The person has no stroke")

```
